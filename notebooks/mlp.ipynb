## Train a Multilayer Perceptron on the MNIST dataset

```python
train, test, validate = load_mnist('/Users/dganguli/src/deeplearning/data')
```

```python
X_train = train[0] # 50,000 X 784
y_train = train[1] # 50,000 X 1
X_train = np.column_stack([X_train, np.ones((len(y_train),1))]) # for the bias 

X_test = test[0]
y_test = test[1]
X_test = np.column_stack([X_test, np.ones((len(y_test),1))])

n_hidden = 16
learning_rate = .1
max_iter = 10000
iter_print = 100

mlp = MultiLayerPerceptron(X_train,n_hidden,y_train)
mlp.fit(learning_rate, max_iter, iter_print)
y_train_hat, scores_train = mlp.predict(X_train)
```

```python
for k in range(25):
    plt.subplot(5,5,k)
    plt.imshow(np.reshape(X_train[k,:-1], (28,28)))
    plt.axis('off')
    plt.title(y_train_hat[k])
```

### Visualize hidden layer weights

```python
for k in range(mlp.W1.shape[0]):
    plt.subplot(4,4,k)
    plt.imshow(np.reshape(mlp.W1[k,:-1], (28,28)))
    plt.axis('off')
```

### Investigate train and test error

```python
train_err = (1 - np.sum((y_train_hat - y_train)==0)/len(y_train))*100

y_test_hat, scores_test = mlp.predict(X_test)
test_err = (1 - np.sum((y_test_hat - y_test)==0)/len(y_test))*100

print 'Train classification error rate: {}%\nTest classification error rate: {}%'.format(train_err,test_err)
```
